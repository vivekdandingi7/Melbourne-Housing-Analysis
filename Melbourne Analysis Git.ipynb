From the data we can observe that, there are several columns where 20% and more than 20% of the data is missing. We know in case of 10% 
we can ignore/ drop missing values but this case we gonna imputate missing values with other values.


```python
sns.boxplot(y=df['Price'])
```

```python
plt.figure(figsize=(4,7))
sns.boxplot(y=df['Price'])
```

```python
plt.figure(figsize=(4,7))
sns.boxplot(y=df['Bedroom'])
```

From the box plot, we can see that there are outliers in price column.If outliers are present ,median is better option
comparing to mean .

```python
df['Price']=df['Price'].fillna(value=df['Price'].median())
df['Car']=df['Car'].fillna(value=df['Car'].median())
df['Bedroom']=df['Bedroom'].fillna(value=2)
df['Bathroom']=df['Bathroom'].fillna(value=1)
df['YearBuilt']=df['YearBuilt'].fillna(value=df['YearBuilt'].median())
df['Landsize']=df['Landsize'].fillna(value=df['Landsize'].median())
```

# Imputed Values:
Price : Median |

Bathroom: 1     |
YearBuilt: mean  |

# Conclusion

Using the Melbourne Housing dataset from Kaggle.com.Some of the conclusions in this analysis could only be applied
to Melbourne Australia, some could be genearlized to similar housing market like Melbourne, some could be applied everywhere.

This conclusion is Melbourne specific,From the data we have the sales price increases as number of rooms,bedrooms,bathrooms increases.When in case of increase in property size we can see a increase in sales price but when we consider the distance factor we see the house sales price decreasing as distance increasing.
 since the southern region of Melbourne is closer to the Central Business District and the beach. Demand is higher in the southern region which causes more apartment constructions that could fit more residents per square feet. 

This conclusion could be generalized to everywhere that property size and the community of the property are important in
price prediction. 
In average,the larger the property and the better the community the property located should mean the higher
the property price
